```python
from transformers import pipeline

# Load a pre-trained sentiment analysis pipeline using DistilBERT
classifier = pipeline('sentiment-analysis')

# Example texts for classification
texts = [
    "I love using the Hugging Face Transformers library!",
    "This movie was terrible and I hated every minute of it.",
    "The weather today is neutral."
]

# Classify the texts
results = classifier(texts)

# Print the results
for i, text in enumerate(texts):
    print(f"Text: '{text}'")
    print(f"Classification: {results[i]['label']}, Score: {results[i]['score']:.4f}\n")
```
